# Netflix Movies & Shows Dataset


This dataset contains information about movies and TV shows available on Netflix.

I selected this dataset because it contains different types of features such as:
- Categorical columns
- Text columns
- Date columns
- Missing values

It is suitable for performing Exploratory Data Analysis (EDA) and learning Pandas commands.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/netflix_data.csv')
df.head()
```

# 7 First-Look Protocol

**C1 Shape**

```python
df.shape
```

The dataset contains rows and columns.
This is a reasonably large dataset for exploratory data analysis and machine learning tasks.

**C2 Data Types**

```python
df.dtypes
```

The dataset contains object, integer, and date-related columns.
Most columns appear correctly typed, but date columns may need conversion.

**C3 Info**

```python
df.info()
```

This command shows the number of non-null values and memory usage.
Several columns contain missing values that may require cleaning.

**C4 Describe**

```python
df.describe()
```

The numerical columns show measures such as mean, median, minimum, and maximum values.
These statistics help identify unusual values and distributions.

**C5 Missing Values**

```python
df.isnull().sum()
```

Some columns contain missing values.
These may need to be dropped or imputed before modeling.

**C6 Value Counts**

```python
df['type'].value_counts()
```

This shows the number of Movies and TV Shows.
It helps understand category balance in the dataset.

**C7 Duplicate Rows**

```python
df.duplicated().sum()
```

This command identifies duplicate records.
Duplicates can negatively affect analysis and model performance.

```python
df.corr(numeric_only=True)
```

**8 Histogram**

```python
df['release_year'].hist(bins=20)
plt.show()
```

**9 GroupBy Mean**

```python
df.groupby('type').mean(numeric_only=True)
```

```python
df['type'].value_counts(normalize=True)*100
```

```python
df.select_dtypes(include='object')
```

```python
df['title'].str.contains('Love', case=False, na=False).sum()
```

Counts titles containing the word "Love".

**13 Sort Values**

```python
df.sort_values('release_year', ascending=False).head()
```

**14 Filter Data**

```python
df[df['release_year'] > 2020]
```

Filters records released after 2020.

**15 Pivot Table**

```python
pd.pivot_table(
    df,
    values='release_year',
    index='type',
    aggfunc='mean'
)
```

**16 Missing Percentage**

```python
(df.isnull().sum()/len(df))*100
```

**19 Kurtosis**

```python
pd.crosstab(df['type'], df['rating'])
```
